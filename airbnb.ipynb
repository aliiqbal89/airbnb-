```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df["Income Band"].value_counts().plot(kind='bar')
```

```python
#examine the distribution of unique caetgories in a categorical colums
```

```python
categorical_cols = df[['BRId','GenderId', "Amount of Credit Cards", "Nationality", "Occupation","Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting","Income Band" ]].columns
for col in categorical_cols:
  print(f"value Count for '{col}':")
  display(df[col].value_counts())
```

```python
##unibraite
#doing it by 1 columns at a time
#column by column
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for i, predictor in enumerate(df[['BRId', 'GenderId', 'Amount of Credit Cards',
                                  'Nationality', 'Occupation', 'Fee Structure',
                                  'Loyalty Classification', 'Properties Owned',
                                  'Risk Weighting', 'Income Band']]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='GenderId')
    plt.title(f'Countplot of {predictor} by Gender')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
##bivariate Analayis
import matplotlib.pyplot as plt
import seaborn as sns

for i, predictor in enumerate(df[['BRId', 'GenderId', 'Amount of Credit Cards',
                                  'Nationality', 'Occupation', 'Fee Structure',
                                  'Loyalty Classification', 'Properties Owned',
                                  'Risk Weighting', 'Income Band']]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Nationality')
    plt.title(f'Countplot of {predictor} by Gender')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col)
    plt.title(f'Countplot of {col}')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)  # optional: rotate labels for better visibility
    plt.tight_layout()
    plt.show()
```

```python
numerical_cols = ['Estimated Income','Superannuation Savings','Amount of Credit Cards','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
# Univariate analysis and visualization
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 3, i+1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.tight_layout()  # Add this to prevent overlapping titles/labels

plt.show()
```

```python
numerical_cols = ['Estimated Income','Superannuation Savings','Amount of Credit Cards','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True,cmap='crest',fmt=".2f")
plt.title("correlation matrix")
plt.show()
```
